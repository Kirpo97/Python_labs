```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
%matplotlib inline

tickers_list = ['GAZP.ME', 'ROSN.ME', 'YNDX.ME']
data = pd.DataFrame(columns=tickers_list)

for ticker in tickers_list:
    data[ticker] = yf.download(ticker,'2016-01-01','2022-08-01')['Adj Close']

((data.pct_change()+1).cumprod()).plot(figsize=(10, 7))

plt.legend(['Газпром', 'Роснефть', 'Яндекс'], title='легенда')
plt.title("Графики изменения акций", fontsize=16)
plt.ylabel('цена (x*100)', fontsize=14)
plt.xlabel('год', fontsize=14)
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```
