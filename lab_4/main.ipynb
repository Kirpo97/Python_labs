# Лабараторная работа №4. ООП

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt 

class Company(): 
            
    def __init__(self):
        self.GAZP = 'GAZP.ME'
        self.ROSN = 'ROSN.ME'
        self.YNDX = 'YNDX.ME'
        self.ADS = 'ADS.DE'
        self.MOEX = 'MOEX.ME'
            
    def __call__(self):
        return self.__dict__
    
    def __str__(self):
        return f"Все тикеты акций: {self.__dict__}"

    def __repr__(self):
        return f"Тикеты энергетического сектора: ['{self.GAZP}', '{self.ROSN}']"            
                
    def draw_plot(self, data1, data2):
        tickers_dict = Company().__dict__
        data = pd.DataFrame(columns=tickers_dict)  
        for ticker in tickers_dict:
            data[ticker] = yf.download(tickers_dict[ticker], data1, data2)['Adj Close']
        ((data.pct_change()+1).cumprod()).plot(figsize=(8, 5))   
        plt.ylabel('цена (x*100)', fontsize=8)
        plt.xlabel('год', fontsize=8)
        plt.grid(
            which="major", 
            color='k', 
            linestyle='-.', 
            linewidth=0.5)
        plt.show() 
        
    def dowload_data(self, data1, data2):
        tickers_dict = Company().__dict__
        data = pd.DataFrame(columns=tickers_dict)  
        for ticker in tickers_dict:
            data[ticker] = yf.download(tickers_dict[ticker], data1, data2)['Adj Close']
        return data   
```

```python
c = Company()
str(c)
```

```python
c.draw_plot('2022-01-01', '2022-12-10')
```
